
##### Read CSV File from Azure Data Lake Storage Account
 CSV Source File Path : "abfss://working-labs@datalakestorageaccountname.dfs.core.windows.net/bronze/daily-pricing/csv"

JSON Target File Path : "abfss://working-labs@datalakestorageaccountname.dfs.core.windows.net/bronze/daily-pricing/json"
##### Spark Methods
- <a href="https://spark.apache.org/docs/latest/sql-getting-started.html#starting-point-sparksession" target="_blank">SparkSession</a>

- <a href="https://spark.apache.org/docs/latest/sql-data-sources-csv.html" target="_blank">DataFrameReader</a>: **`csv`**,  **`option (header,separator)`** ,  **`schema`**

- <a href="https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/data_types.html" target="_blank">SparkDataTypes</a>: **`ArrayType`**, **`DoubleType`**, **`IntegerType`**, **`LongType`**, **`StringType`**, **`StructType`**, **`StructField`**

- <a href="https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.types.StructType.html" target="_blank">StructType</a>


- <a href="https://spark.apache.org/docs/3.1.2/api/python/reference/api/pyspark.sql.DataFrameWriter.json.html" target="_blank">DataFrameWriter</a>: **`json`**,  **`mode (overwrite,append)`** 

```python
#sourceCSVFilePath = 'abfss://working-labs@adlsudadatalakehousedev.dfs.core.windows.net/bronze/daily-pricing/csv'
sourceCSVFilePath = '/Volumes/databricks_project/rgr_storage/working_labs/bronze/daily-pricing/csv/'
#targetJSONFilePath = 'abfss://working-labs@adlsudadatalakehousedev.dfs.core.windows.net/bronze/daily-pricing/json'
targetJSONFilePath = '/Volumes/databricks_project/rgr_storage/working_labs/bronze/daily-pricing/json'
```

```python
sourceCSVFileDF = ( spark.
                   read.
                   option("header","true").
                   csv(sourceCSVFilePath) 
                   )
```

```python
display (sourceCSVFileDF )
```

```python
from pyspark.sql.types import *
sourceCSVFileSchema = StructType([
  StructField("DATE_OF_PRICING",StringType(),True),
  StructField("ROW_ID",IntegerType(),True),
  StructField("STATE_NAME",StringType(),True),
  StructField("MARKET_NAME",StringType(),True),
  StructField("PRODUCTGROUP_NAME",StringType(),True),
  StructField("PRODUCT_NAME",StringType(),True),
  StructField("VARIETY",StringType(),True),
  StructField("ORIGIN",StringType(),True),
  StructField("ARRIVAL_IN_TONNES",DecimalType(10,2),True),
  StructField("MINIMUM_PRICE",StringType(),True),
  StructField("MAXIMUM_PRICE",StringType(),True),
  StructField("MODAL_PRICE",StringType(),True),
  ])
```

```python
sourceCSVFileDF = ( spark.read\
                   .option("header", True)\
                   .schema(sourceCSVFileSchema)\
                   .csv(sourceCSVFilePath) 
                  )
```

```python
display (sourceCSVFileDF )
```

```python
sourceCSVFileDF.printSchema
```

```python
(
    sourceCSVFileDF.write.mode("overwrite")
    .option("compression", "gzip")
    .option("dateFormat", "yyyy-MM-dd")
    .option("timestampFormat", "yyyy-MM-dd'T'HH:mm:ss.SSS'Z'")
    .option("ignoreNullFields", True)
    .option("charset", "UTF-8")
    .json(targetJSONFilePath)
)
```

```python
(
    sourceCSVFileDF.write.mode("overwrite")
    .option("header", True)
    .option("sep", ",")
    .option("quote", '"')
    .option("escape", '"')
    .option("nullValue", "NULL")
    .option("dateFormat", "yyyy-MM-dd")
    .option("timestampFormat", "yyyy-MM-dd HH:mm:ss")
    .csv(sourceCSVFilePath)
)
```
