# Create and Manage Clusters

- <a href="https://docs.databricks.com/en/compute/index.html">**Databricks Clusters**</a>


A Databricks cluster is a set of computation resources and configurations to run the Data Analytical Code that we develop in the Notebook

There are 5 Type Of the Clusters are available and see below table for their purpose and usage
| Cluster Type | Usage |
| --- | --- |
|All-purpose compute| Running Interactive Workloads and Code Development Purposes|
|Job compute| Running Production Workloads and Automatically Start/Stop by the Code Using|
|SQL Warehouse| Running only SQL Workloads|
|Vector Search| Running Machine Learning Models(LM)|
|Pools| Ready to Use Idle Instances to quickly run the code reducing start time|




## Create Cluster

Steps:
1. Use the left sidebar to navigate to the **Compute** page by clicking on the **Compute** icon
1. Click the blue **Create Compute** button
1. For the **Cluster name**, use "Single_Node_Dev_Cluster" 
1. Set the **Cluster mode** to **Single Node** (This cluster charges minimum amount)
1. Choose **Databricks runtime version** as one of the LTS(Long Time Support) 14.3 LTS version
1. Choose *Worker Node Type** as one of the DS3_v2 or DS4_v2 under General Purpose Family of Clusters and Small Size is enough to run Dev Workloads
1. Leave boxes checked for the default settings under the **Photon Accelaration**
1. Chnage the Terminate After Settings to 10 Minutes as automatically cluster shutdown if we are not using longer than 10 minutes
1. Click the blue **Create Cluster** button

It Takes Couple of Minites To Complet
