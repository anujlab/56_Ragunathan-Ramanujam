```python
#sourceCSVFilePath = 'abfss://working-labs@adlsudadatalakehousedev.dfs.core.windows.net/bronze/daily-pricing/csv'
sourceCSVFilePath = '/Volumes/databricks_project/rgr_storage/working_labs/bronze/daily-pricing/csv/'
#sourceJSONFilePath = 'abfss://working-labs@adlsudadatalakehousedev.dfs.core.windows.net/bronze/daily-pricing/json'
sourceJSONFilePath = '/Volumes/databricks_project/rgr_storage/working_labs/bronze/daily-pricing/json/'
```

```python
sourceJSONFileDF = spark.read.json(sourceJSONFilePath)
display(sourceJSONFileDF)
```

```python
sourceJSONFileSchemaDDL = "DATE_OF_PRICING string,ROW_ID integer,STATE_NAME string,MARKET_NAME string,PRODUCTGROUP_NAME string,PRODUCT_NAME string,VARIETY string,ORIGIN string,ARRIVAL_IN_TONNES Decimal(18,2),MINIMUM_PRICE string,MAXIMUM_PRICE string,MODAL_PRICE string"

display(sourceJSONFileDF)
```

```python
sourceJSONFileDF = spark.read.schema(sourceJSONFileSchemaDDL).json(sourceJSONFilePath)
```

```python
df = sourceJSONFileDF.select("PRODUCT_NAME")
display(df)
```

```python
(
  sourceJSONFileDF.
  drop("ROW_ID")
)
```

```python
(
  sourceJSONFileDF.
   select("PRODUCT_NAME","ARRIVAL_IN_TONNES").
   filter("ARRIVAL_IN_TONNES > 100")
)
```

```python
from pyspark.sql.functions import col
(sourceJSONFileDF.
 select("PRODUCT_NAME","ARRIVAL_IN_TONNES").
 filter("ARRIVAL_IN_TONNES > 100").
 where(col("STATE_NAME") == "Andhra Pradesh")
)
```

```python
(sourceJSONFileDF.
 select("PRODUCT_NAME","ARRIVAL_IN_TONNES").
 filter("ARRIVAL_IN_TONNES > 100").
 where(col("STATE_NAME") == "Andhra Pradesh").
 sort("ARRIVAL_IN_TONNES")
)
```

```python
from pyspark.sql.functions import desc
(sourceJSONFileDF.
 select("PRODUCT_NAME","ARRIVAL_IN_TONNES").
 filter("ARRIVAL_IN_TONNES > 100").
 where(col("STATE_NAME") == "Andhra Pradesh").
 sort(desc("ARRIVAL_IN_TONNES"))
)
```
